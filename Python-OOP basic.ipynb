#### First-Class Functions in Python
<small> A language has first-class functions if it treats functions as "first-class citizens". This means functions are objects that can be passed as arguments, returned from other functions, and assigned to variables
<small>

```python
# Example: Assigning a function to a variable
def square(x):
    return x * x

f = square  # No parentheses; we are assigning the function object
print(f(5)) # Output: 25 [5, 6]
```

#### Higher-Order Functions
<small> A function is "higher-order" if it accepts other functions as arguments or returns a function as its result
<small>

```python
def my_map(func, arg_list):
    result = []
    for i in arg_list:
        result.append(func(i)) # Execute the passed-in function
    return result

def cube(x):
    return x * x * x
```

```python
my_map(cube, [1, 2, 3, 4])
```

<h4>Closures in Python</h4>

<small>A closure is an inner function that <b>"remembers"</b> and has access to variables in the local scope where it was created, even after the outer function has finished executing. These "remembered" variables are called <b>free variables.</b></small>

<small>To form a closure, three conditions must be met:</small>

<small>

1. There must be a **nested function** (a function inside a function)
2. The inner function must **refer to a variable** defined in the outer function
3. The outer function must **return the inner function**

</small>

```python
# Example: A simple closure
def outer_func(msg):
    message = msg # 'message' is a local variable to outer_func

    def inner_func():
        print(message) # 'message' is a free variable to inner_func [11, 13]

    return inner_func # Return the function object without executing it [12]

hi_func = outer_func('Hi')
hello_func = outer_func('Hello')
```

```python
hi_func()
hello_func()
```

```python
def make_interest_calculator(rate):
    def calculate(principal, years):
        return principal * (1 + rate) ** years
    return calculate

savings_calc   = make_interest_calculator(0.04)
```

```python
savings_calc(1000, 5)
```

```python
#decorators
def decorator(func):
    def wrapper():
        print("Something is happening before the function is called.")
        func()
        print("Something is happening after the function is called.")
    return wrapper  
```
