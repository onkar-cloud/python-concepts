#### O(1) — Constant Time

```python
def get_first_element(arr):
    return arr[0]  # Always one operation, regardless of array size
```

```python
arr = [10, 20, 30, 40, 50]
print(get_first_element(arr))  # 10
```

#### O(1) — Constant Time

```python
def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    counter = 0  # To count the number of operations
    while low <= high:
        counter += 1  # Increment the operation count
        mid = (low + high) // 2
        if arr[mid] == target:
            return f"index of the candidate number for search: {mid},\nnumber of operations performed: {counter}"  # Return index and operation count
        elif arr[mid] < target:
            low = mid + 1      # Discard left half
        else:
            high = mid - 1     # Discard right half
    return -1, f"number of operations performed: {counter}"  # Return -1 and operation count if not found
```

```python
import numpy as np
arr = np.random.randint(1,1000, size=100).tolist()  # Generate a random list of 100 integers between 1 and 99
arr = sorted(arr)  # Binary search requires sorted array
print(f"array for binary search: {arr}")
random_number = np.random.choice(arr)
print(f"random number to search in array: {random_number}")
print(binary_search(arr, random_number))   # 3  (index)
# n=7 → at most 3 steps (log₂7 ≈ 3)
```
