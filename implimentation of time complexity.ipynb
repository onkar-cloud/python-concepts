<h4>What is Time Complexity?</h4>
<small>Time complexity describes how the runtime of an algorithm grows relative to the size of its input (n).

It's expressed using Big O notation, which captures the worst-case growth rate, ignoring constants and lower-order terms.</small>


#### O(1) — Constant Time
<small> No matter how large the input, the operation always takes the same time.<small>

```python
def get_first_element(arr):
    return arr[0]  # Always one operation, regardless of array size
```

```python
arr = [10, 20, 30, 40, 50]
print(get_first_element(arr))  # 10
```

<h4>O(log n) — Logarithmic Time</h4>

<small>The input is halved (or divided) at each step. Very efficient for large inputs.</small>

```python
def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    counter = 0  # To count the number of operations
    while low <= high:
        counter += 1  # Increment the operation count
        mid = (low + high) // 2
        if arr[mid] == target:
            return f"index of the candidate number for search: {mid},\nnumber of operations performed: {counter}"  # Return index and operation count
        elif arr[mid] < target:
            low = mid + 1      # Discard left half
        else:
            high = mid - 1     # Discard right half
    return -1, f"number of operations performed: {counter}"  # Return -1 and operation count if not found
```

```python
import numpy as np
arr = np.random.randint(1,1000, size=100).tolist()  # Generate a random list of 100 integers between 1 and 99
arr = sorted(arr)  # Binary search requires sorted array
print(f"array for binary search: {arr}")
random_number = np.random.choice(arr)
print(f"random number to search in array: {random_number}")
print(binary_search(arr, random_number))   # 3  (index)
# n=7 → at most 3 steps (log₂7 ≈ 3)
```

<h4>O(n) — Linear Time</h4>

<small>Runtime grows directly proportional to input size. You touch each element once.</small>

```python
def find_max(arr):
    max_val = arr[0]
    for num in arr:        # Visits every element once
        if num > max_val:
            max_val = num
    return max_val
```

```python
arr = [3, 1, 4, 1, 5, 9, 2, 6]
print(find_max(arr))
```

<h4>O(n log n) — Linearithmic Time</h4>
<small>Common in efficient sorting algorithms. Each of the n elements is processed log n times.</small>

```python
def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    
    return merge(left, right)

def merge(left, right):
    result = []
    i = j = 0
    
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    
    result.extend(left[i:])
    result.extend(right[j:])
    return result
```

```python

print(merge_sort([5, 2, 9, 8, 1, 5, 6]))
```

<h4> O(n²) — Quadratic Time</h4>
<small>A loop inside a loop. For every element, you process every other element.</small>

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr
```

```python
sorted = bubble_sort([5, 2, 9, 8, 1, 5, 6])
print(f"Sorted array: {sorted}")
print(f"Bubble sort is of order n^2 because of the nested loops. The outer loop runs n times and the inner loop runs n-i-1 times, leading to a total of approximately n*(n-1)/2 comparisons in the worst case, which simplifies to O(n^2).")
```

<h4>O(n³) — Cubic Time</h4>
<small>Three nested loops. Rare in practice; usually signals a need for optimization.</small>

```python
def matrix_multiply(A, B):
    n = len(A)
    C = [[0] * n for _ in range(n)]
    
    for i in range(n):
        for j in range(n):
            for k in range(n):
                C[i][j] += A[i][k] * B[k][j]
    
    return C
```

```python
A = [[1, 2], 
     [3, 4]]
B = [[5, 6], 
     [7, 8]]
print(matrix_multiply(A, B))
```

<h4>O(2ⁿ) — Exponential Time</h4>
<small>each additional input element multiplies the runtime rather than just adding or scaling it linearly.</small>

```python
def fib(n):
    if n <= 1:
        return n
    return fib(n-1) + fib(n-2)
```

```python
print(f"fib(1) = {fib(1)}")
print(f"fib(2) = {fib(2)}")
print(f"fib(3) = {fib(3)}")
print(f"fib(4) = {fib(4)}")
print(f"fib(5) = {fib(5)}")
print(f"fib(6) = {fib(6)}")
print(f"fib(7) = {fib(7)}")
print(f"fib(8) = {fib(8)}")
print(f"fib(9) = {fib(9)}")
print(f"fib(10) = {fib(10)}")
```
